```python
# sigmoid 함수
import numpy as np

def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
# MSE 함수
def MSE(tar, val):
  if not val:
    return False
  
  error = 0
  for e in range(len(tar)):
    error += 0.5 * (tar[e]-val[e])**2
  
  return error
```

```python
x1 = 0.2
x2 = 0.5

tar1 = 0.2
tar2 = 0.7

weight0 = [[0.1, 0.2], [0.3, 0.1]]
weight1 = [[0.4, 0.5], [0.1, 0.3]]

learning_rate = 0.3
iter = 1000

def update_l1_weight(tar1, tar2, y1, y2, w1, w2, a, updW):
  e1 = -(tar1-y1)*y1*(1-y1)*w1
  e2 = -(tar2-y2)*y2*(1-y2)*w2
  v1 = a*(1-a)
  v2 = a
  df = (e1+e2)*v1*v2

  return updW - learning_rate * df

def update_l2_weight(tarY, y, predY, updW):
  v1 = -(tarY-y)
  v2 = y*(1-y)
  df = v1*v2*predY

  return updW - learning_rate * df
```

```python
for i in range(iter):
  z10 = x1 * weight0[0][0] + x2 * weight0[1][0];
  a10 = sigmoid(z10);
  z11 = x1 * weight0[0][1] + x2 * weight0[1][1];
  a11 = sigmoid(z11);

  z20 = a10 * weight1[0][0] + a11 * weight1[1][0];
  a20 = sigmoid(z20);
  z21 = a10 * weight1[0][1] + a11 * weight1[1][1];
  a21 = sigmoid(z21);

  e_tar = MSE([tar1, tar2], [a20, a21])
  
  print(f'{i+1}th iteration : y1 = {a20}, y2 = {a21}, E = {e_tar}')

  nw0 = [
    [
      update_l1_weight(tar1, tar2, a20, a21, weight1[0][0], weight1[0][1], a10, weight0[0][0]),
      update_l1_weight(tar1, tar2, a20, a21, weight1[1][0], weight1[1][1], a11, weight0[0][1]),],
    [
      update_l1_weight(tar1, tar2, a20, a21, weight1[0][0], weight1[0][1], a10, weight0[1][0]),
      update_l1_weight(tar1, tar2, a20, a21, weight1[1][0], weight1[1][1], a11, weight0[1][1]),]
    
  ]
  nw1 = [
    [update_l2_weight(tar1, a20, a10, weight1[0][0]), update_l2_weight(tar2, a21, a10, weight1[0][1])],
    [update_l2_weight(tar1, a20, a11, weight1[1][0]), update_l2_weight(tar2, a21, a11, weight1[1][1])]
  ]


  i, j= 0, 0
  for v in nw0:
    for vv in v:
      if i > 1:
        break
      weight0[i][j] = vv
      i+=1
    j+=1

  i, j= 0, 0
  for v in nw1:
    for vv in v:
      if i > 1:
        break
      weight1[i][j] = vv
      i+=1
    j+=1
    
print(f'target 1 : {tar1}, target 2 : {tar2}')
```
